```python
import json
import requests
import glob
import pickle
from tqdm import tqdm
from xml.etree import ElementTree

url = "http://localhost:9003/?output_format=jsonl"
files = glob.glob("./train_texts/*.txt")
out_files = './out_file/{}.pkl'

verbs = ["fin", "aglt", "praet", "impt", "imps", "inf", "pcon", "pant", "ger", "pact", "ppas"]
nouns = ["subst", "depr"]
adjectives = ["adj", "adja", "adjp", "adjc"]
```

```python
files
```

```python
def parse_line(line):
    r = requests.post(url, data = line.encode('utf-8'))
    out = json.loads(r.text)
    return out

flatten = lambda t: [item for sublist in t for item in sublist]
```

```python
files_text = []
for file in files:
    with open(file) as f:
        files_text.append((file.split('/')[-1].split('.')[0], f.read().replace('\n', '')))
```

```python
flatten(parse_line(files_text[0][1]))
```

```python

matched = 0
miss_matched = 0
CTAG_POSSITION = 2
BASE_POSSITION = 1
TEXT_POSSITION = 0
for name, text in tqdm(files_text):
    output = {
        "noun": {
            
        },
        "verb": {
            
        },
        "adjective": {
            
        }
    }
    parsed = flatten(parse_line(text))
    for word in parsed:
        wordType = word[CTAG_POSSITION].split(':')[0]
        
        if wordType in verbs:
            if word[1] in output['verb']:
                output['verb'][word[1]] += 1
            else:
                output['verb'][word[1]] = 1
                
        if wordType in nouns:
            if word[1] in output['noun']:
                output['noun'][word[1]] += 1
            else:
                output['noun'][word[1]] = 1
                
        if wordType in adjectives:
            if word[1] in output['adjective']:
                output['adjective'][word[1]] += 1
            else:
                output['adjective'][word[1]] = 1
                
    pickle.dump( output, open( out_files.format(name), "wb" ) )
    
```
